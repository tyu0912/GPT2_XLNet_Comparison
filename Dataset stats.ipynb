```python
import os
```

```python
stats = {}
overall_words = []

for root, dirs, files in os.walk("CorpusOfPresidentialSpeeches", topdown=False):
    if '\\' in root:
        rootf, pres = root.split('\\')
        stats[pres] = [0,0,0]
        words = []

        for name in files:
            speech = os.path.join(root, name)
            stats[pres][0] += 1 

            with open(speech, 'r', encoding="utf-8") as f:
                text = f.read()
                text = text.replace('\n', '')
                words += text.split(' ')
                overall_words += text.split(' ')
                #print(text)

        stats[pres][1] = len(words)
        stats[pres][2] = len(set(words))
        
    
    stats['overall'] = [0, len(overall_words), len(set(overall_words))]
                             
for k,v in stats.items():
    print (k, v)
```
