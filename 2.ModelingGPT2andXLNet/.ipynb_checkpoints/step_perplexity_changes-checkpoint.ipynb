```python
from matplotlib import pyplot as plt
import os
import re
```

```python
data = {}

for root, dirs, files in os.walk("gpt2_tuned", topdown=False):
    for name in files:
        if 'eval_results' in name:
            checkpoint = root[root.find("-")+1:]
            
            try:
                checkpoint = int(checkpoint)
            except:
                checkpoint = 800
                
            fh = open(os.path.join(root, name)).read()
            perplexity = float(re.findall("\d+\.\d+", fh)[0])
            
            data[checkpoint] = perplexity

print(data)
```

```python
plt.scatter(data.keys(), data.values())
plt.xlabel("steps")
plt.ylabel("perplexity")
```
