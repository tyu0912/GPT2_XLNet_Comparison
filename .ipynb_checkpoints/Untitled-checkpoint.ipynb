```python
out = open("all_obama.txt","w+")

for root, dirs, files in os.walk("CorpusOfPresidentialSpeeches/obama", topdown=False):
    for file in files[2:]:
        path = root + '/' + file

        fh = open(path, "r").read()
        fh = re.sub(r"\<[^\>]*\>", '', fh)
        fh = fh.strip()
        
        
        doc = nlp(fh)
        mapping = sorted(list(set([(X.text, X.label_) for X in doc.ents])), key=lambda x: -len(x[0]))
        
        for maps in mapping:
            if maps[1] not in ["CARDINAL", "GPE"]:
                fh = fh.replace(maps[0], "<"+maps[1]+">")
        
        fh = fh.split(".")
        fh = list(map(lambda x: x.replace('\n',' '), fh))
        
        for l in fh:
            out.write("<ss>" + l.strip() + '</ss>\n')
        
        #for item in fh.splitlines()[2:]:
            
        #    print(item)
        #    break
            
            #for line in item.split('.')[:-1]:
            #    print(line.strip())
```
