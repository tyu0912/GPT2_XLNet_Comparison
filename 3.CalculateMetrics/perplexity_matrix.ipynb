```python
import os
import statistics
from collections import defaultdict
import pandas as pd
```

```python
files = ["test_gpt2_results", "test_gpt2tuned_results", "test_xlnet-base-cased_results",
         'LSTM-attention-result-new-dim2', 'gpt2_BERT_result-new-dim', 'original']

model_paths=["gpt2", "gpt2_tuned", "xlnet-base-cased"]
```

```python
def calculate_mean(f, mp):

    path = f"20191129_run/{f}_{mp}_perplexity.txt"
        
    fh = open(path).read().split("<new_line>")
    all_perp = []

    for line in fh[1:]:
        perplexity = line.split("<entry>")[3]
        all_perp.append(float(perplexity))

    return statistics.mean(all_perp)
    
    #print(statistics.stdev(all_perp))
```

```python
def default_to_regular(d):
    if isinstance(d, defaultdict):
        d = {k: default_to_regular(v) for k, v in d.items()}
    return d
```

```python
results = defaultdict(lambda: defaultdict(float))

for f in files:
    for mp in model_paths:
        
        try:
            results[f][mp] = calculate_mean(f, mp)
        except:
            pass
        
```

```python
df = pd.DataFrame(default_to_regular(results)).T
```

```python
df = df.rename(index={'test_gpt2_results': "GPT2 Untuned", 
                 'test_gpt2tuned_results': "GPT2 Tuned", 
                 'test_xlnet-base-cased_results': "XLNet",
                 'LSTM-attention-result-new-dim2': "LSTM-Attention",
                 'gpt2_BERT_result-new-dim': "BERT Smash",
                }, 
         columns={'gpt2': "GPT2 Untuned",
                  'gpt2_tuned': "GPT2 Tuned",
                  'xlnet-base-cased': "XLNet"
                 }
         )
```

```python
df.index.name = "Sentence Generation Model"
df.columns.name = "Validation Model"
```
