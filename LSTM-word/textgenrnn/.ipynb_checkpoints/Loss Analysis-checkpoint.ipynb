```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
training = open("textgenrnn_loss.txt").read().splitlines()
```

```python
losses = []

for data in training:
    if 'loss' in data:
        loss = float(data[70:])
        
        losses.append(loss)
```

```python
x = range(len(losses))
plt.scatter(x, losses)
```
