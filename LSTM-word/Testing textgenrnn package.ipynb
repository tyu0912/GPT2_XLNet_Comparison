```python
def convert_file2lines(in_path, out_path):
    file = open(in_path).read()
    file = file.replace("<speech_sep>", "")
    
    file = file.split('.')
    
    out = open(out_path, 'w')
    
    for line in file:
        sentence = line + '.\n'
        out.write(sentence)
        
train = "../1.DataPreparationResults/combined_data/train.txt"
val = "../1.DataPreparationResults/combined_data/val.txt"
test = "../1.DataPreparationResults/combined_data/test.txt"

convert_file2lines(train, "train_sentence_split.txt")
convert_file2lines(val, "val_sentence_split.txt")
convert_file2lines(test, "test_sentence_split.txt")
```
