```python
generator = tgr()
```

```python
generator.train_from_file('train.txt', header=False, is_csv=True, new_model=True, num_epochs=30, word_level=True, max_length=100)
```

```python
result = generator.generate(1, prefix="Through its mysterious channels we come to wider acquaintance with surroundings and men.", return_as_list = True, top_n=1, temperature=[0.9])
print(result)
```

```python
fh = open("../LSTM-test-file.txt").read().splitlines()
output = open("LSTM-attention-result-test-fixed-30-epoch.txt", 'w')
```

```python
for i,line in enumerate(fh):
    #print("###############################")
    #print(line)
    try:
        prediction = generator.generate(1, prefix=line, return_as_list = True, top_n=1, temperature=[0.9])
        prediction = prediction[0].replace('\n','').split('.')[1].strip()
        prediction = prediction = prediction[:1].upper()  + prediction[1:] + '.'
        actual = fh[i+1]

        output.write("<new_line>" + line + '<entry>' + actual + '<entry>' + prediction + '<end_line>')
        
    except:
        pass
    #print(prediction)
    #print(actual)
```
