```python
from textgenrnn.textgenrnn import textgenrnn as tgr
```

```python
generator = tgr()
```

```python
generator.train_from_largetext_file('../textgenrnn/train_sentence_split.txt',new_model=True,num_epochs=30,word_level=True, max_length=30)
```

```python
result = generator.generate(1, prefix="This room from which I speak was the scene of work and accomplishment of our Presidents for over a century.", return_as_list = True, top_n=1, temperature=[0.9])
```

```python
parsed_result = result[0].replace('\n','').split('.')[0].strip()
```

```python
parsed_result = parsed_result[:1].upper()  + parsed_result[1:] + '.'
```

```python
fh = open("../LSTM-test-file.txt").read().splitlines()
output = open("LSTM-attention-result-new-dim.txt", 'w')
```

```python
for i,line in enumerate(fh):
    print("###############################")
    print(line)
    prediction = generator.generate(1, prefix=line, return_as_list = True, top_n=1, temperature=[0.9])
    prediction = prediction[0].replace('\n','').split('.')[0].strip()
    prediction = prediction = prediction[:1].upper()  + prediction[1:] + '.'
    actual = fh[i+1]
    
    output.write("<new_line>" + line + '<entry>' + actual + '<entry>' + prediction + '<end_line>')
    #print(prediction)
    #print(actual)
```
